```python
# Questão 3, análise exploratória do dataset Titanic
# Abrindo o arquivo e selecionando as colunas de interesse
import numpy as np
import pandas as pd
from scipy import stats

# Colunas de interesse - atributos
att = ['PassengerId','Survived','Pclass','Sex','Age','SibSp','Fare']

data = pd.read_csv('titanic.csv', usecols=att)
data
```

```python
# Análise para Tipo de Passagem
Pclass = data['Pclass']

# Média
avrg = np.average(Pclass)
print("Media: ", avrg)

# Média podada
trimmed = stats.trim_mean(Pclass, proportiontocut=0.20)
print("Trimmed: ", trimmed)

# Winsorização
wins = np.average(stats.mstats.winsorize(Pclass, limits=0.20))
print("Wins: ", wins)

# Mediana
median = np.median(Pclass)
print("Mediana: ", median)

# Moda
mode = stats.mode(Pclass)
print("Moda: ", mode.mode)
```

    Média, moda, mediana e as médias podadas ajudam a entender qual a condição financeira da maioria dos passageiros do navio, uma vez que o atributo Pclass determina a classe de passagem comprada. Pclass assume os valores 1, 2 e 3 para primeira, segunda e terceira classe, respectivamente.
    Dessa forma, pode-se perceber que existiam mais passageiros de segunda e terceira classes em relação aos passageiros de primeira classe.

```python
# Análise para gênero
sex = data['Sex']

# Moda
mode = stats.mode(sex)
print("Moda: ", mode.mode)
```

    Para o atributo de gênero (sex), a medida que faz mais sentido é a moda. Através da análise desse atributo pode ser visto que existiam mais homens do que mulheres no navio.

```python
# Análise para Idade
age = data['Age'][~np.isnan(data['Age'])] # Seleciona apenas os valores de idade diferentes de NaN

# Média
avrg = np.average(age)
print("Media: ", avrg)

# Média podada
trimmed = stats.trim_mean(age, proportiontocut=0.20)
print("Trimmed: ", trimmed)

# Winsorização
wins = np.average(stats.mstats.winsorize(age, limits=0.20))
print("Wins: ", wins)

# Mediana
median = np.median(age)
print("Mediana: ", median)

# Moda
mode = stats.mode(age)
print("Moda: ", mode.mode)
```

    Nós achamaos que a idade também pode ser um fator determinante de sobrevivência dos passageiros, uma vez que pode ter sido determinada preferência de salvamento das pessoas conforme sua idade. Através da observação dos valores acima podemos dizer que as pessoas do navio tinham aproximadamente 30 anos.

```python
# Análise para número de irmãos / cônjuges
sib = data['SibSp'];

# Média
avrg = np.average(sib)
print("Media: ", avrg)

# Média podada
trimmed = stats.trim_mean(sib, proportiontocut=0.20)
print("Trimmed: ", trimmed)

# Winsorização
wins = np.average(stats.mstats.winsorize(sib, limits=0.20))
print("Wins: ", wins)

# Mediana
median = np.median(sib)
print("Mediana: ", median)

# Moda
mode = stats.mode(sib)
print("Moda: ", mode.mode)
```

    Nesse caso é possível observar que a maioria das pessoas não possuiam irmãos / cônjuges a bordo. Esse fator também pode ser importante para determinar se uma pessoa sobrevive ou não.

```python
# Análise para Tarifa
fare = data['Fare'];

# Média
avrg = np.average(fare)
print("Media: ", avrg)

# Média podada
trimmed = stats.trim_mean(fare, proportiontocut=0.20)
print("Trimmed: ", trimmed)

# Winsorização
wins = np.average(stats.mstats.winsorize(fare, limits=0.20))
print("Wins: ", wins)

# Mediana
median = np.median(fare)
print("Mediana: ", median)

# Moda
mode = stats.mode(fare)
print("Moda: ", mode.mode)
```

    Assim como no caso da classe da passagem, nós supomos que as tarifas que o passageiro pagou poderiam ter afetado no seu tratamento quando ocorreu a emergência. 
    Nesse caso, observamos uma certa diferença entre os valores da média comum para as médias podadas. Isso ocorre devido a presença de valores discrepantes ("outliers") de tarifas pagas por alguns poucos passageiros.

            B) Existência de água filtrada (AGUA_FILTRADA)
                Tipo de dado: nominal

            C) Inexistência de energia elétrica (ENERGIA_INEXISTENTE)
                Tipo de dado: nominal

            D) Inexistência de esgoto (ESGOTO_INEXISTENTE)
                Tipo de dado: nominal
                Operações: moda, porcentagem, igual (==), diferente (!=)

            E) Existência de alimentação (ALIMENTACAO)
                Tipo de dado: nominal
                Operações: moda, porcentagem, igual (==), diferente (!=)

       Os atributos acima foram escolhidos pois conseguimos os dados necessários para responder a perguntas sobre o oferecimento de serviços básicos às escolas de São Paulo.

```python
# Questão 3, análise exploratória do dataset Instituições de Ensino
# Abrindo o arquivo e selecionando as colunas de interesse
import numpy as np
import pandas as pd
from scipy import stats

att = ['ALIMENTACAO', 'AGUA_FILTRADA', 'AGUA_INEXISTENTE', 'ENERGIA_INEXISTENTE', 'ESGOTO_INEXISTENTE']

# O arquivo está em unicode e os separadores são ';', portanto é preciso mandar essa informação à função
data = pd.read_csv('escolas.csv', usecols=att, sep=';', encoding='ISO-8859-1')

# Removendo linhas inválidas
data = data.dropna()

data
```

```python
# Análise para Alimentação
alimentacao = data['ALIMENTACAO']
n_sim = 0
n_nao = 0

# Moda
mode = stats.mode(alimentacao)
print("Moda: ", mode.mode)

# Análise do número de Sim e Não e suas porcentagens
for x in alimentacao:
    if x == 'Sim':
        n_sim = n_sim + 1
    else:
        n_nao = n_nao + 1

print("Número de Sim: ", n_sim)
print("Número de Não: ", n_nao)
print("Porcentagem de Sim: ", n_sim/(n_sim + n_nao))
print("Porcentagem de Não: ", n_nao/(n_sim + n_nao))
```

```python
# Análise para Água Filtrada
agua_filtrada = data['AGUA_FILTRADA']
n_filtrada = 0
n_nao_filtrada = 0

# Moda
mode = stats.mode(agua_filtrada)
print("Moda: ", mode.mode)

# Análise do número de Sim e Não e suas porcentagens
for x in agua_filtrada:
    if x == 'Filtrada':
        n_filtrada = n_filtrada + 1
    elif x == 'Não filtrada':
        n_nao_filtrada = n_nao_filtrada + 1

print("Número de Filtrada: ", n_filtrada)
print("Número de Não filtrada: ", n_nao_filtrada)
print("Porcentagem de Filtrada: ", n_filtrada/(n_filtrada + n_nao_filtrada))
print("Porcentagem de Não filtrada: ", n_nao_filtrada/(n_filtrada + n_nao_filtrada))
```

```python
# Análise para Água
agua = data['AGUA_INEXISTENTE']
n_sim = 0
n_nao = 0

# Moda
mode = stats.mode(agua)
print("Moda: ", mode.mode)

# Análise do número de Sim e Não e suas porcentagens
for x in agua:
    if x == 'Sim':
        n_sim = n_sim + 1
    else:
        n_nao = n_nao + 1

print("Número de Sim: ", n_sim)
print("Número de Não: ", n_nao)
print("Porcentagem de Sim: ", n_sim/(n_sim + n_nao))
print("Porcentagem de Não: ", n_nao/(n_sim + n_nao))
```

```python
# Análise para Energia
energia = data['ENERGIA_INEXISTENTE']
n_sim = 0
n_nao = 0

# Moda
mode = stats.mode(energia)
print("Moda: ", mode.mode)

# Análise do número de Sim e Não e suas porcentagens
for x in energia:
    if x == 'Sim':
        n_sim = n_sim + 1
    else:
        n_nao = n_nao + 1

print("Número de Sim: ", n_sim)
print("Número de Não: ", n_nao)
print("Porcentagem de Sim: ", n_sim/(n_sim + n_nao))
print("Porcentagem de Não: ", n_nao/(n_sim + n_nao))
```

```python
# Análise para Esgoto
esgoto = data['ESGOTO_INEXISTENTE']
n_sim = 0
n_nao = 0

# Moda
mode = stats.mode(esgoto)
print("Moda: ", mode.mode)

# Análise do número de Sim e Não e suas porcentagens
for x in esgoto:
    if x == 'Sim':
        n_sim = n_sim + 1
    else:
        n_nao = n_nao + 1

print("Número de Sim: ", n_sim)
print("Número de Não: ", n_nao)
print("Porcentagem de Sim: ", n_sim/(n_sim + n_nao))
print("Porcentagem de Não: ", n_nao/(n_sim + n_nao))
```
